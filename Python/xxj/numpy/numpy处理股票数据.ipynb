```python
import numpy as np
import datetime#处理日期和时间

def datestr2num(s):
    # 星期一 0 …… 星期六 5 星期日 6
    # 从csv读出是bytes数据，需要转换为str
    #解析字符串当中的日期
    return datetime.datetime.strptime(s.decode('utf-8'), '%m/%d/%Y').weekday()#utf8解码 month day Year
#对数据进行转化 

dates, c = np.loadtxt('Apple.csv', delimiter=',', usecols=(0, 1), unpack=True, converters={0: datestr2num})
#第0列用datestr2num函数做转换 取出第0 1列
averages = np.zeros(5)
for i in range(5):
    index = np.where(dates == i)
    prices = np.take(c, index)     # 类似数组索引 读取索引的值
    avg = np.mean(prices)
    averages[i] = avg
    print("Day {} prices: {},avg={}".format(i, prices, avg))

top = np.max(averages)
top_index = np.argmax(averages)
bot = np.min(averages)
bot_index = np.argmin(averages)
print('highest:{}, top day is {}'.format(top, top_index)) #平均收盘价最高
print('lowest:{},bottom day is {}'.format(bot, bot_index)) #平均收盘价最低
```
