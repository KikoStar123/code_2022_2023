```python
data=pd.read_csv("data6.csv")
print(data)
data['时间']=pd.to_datetime(
    data.注册时间,
    format="%Y/%m/%d"
    )#转换为标准日期格式
data['年月']=data.时间.dt.strftime("%Y-%m")
data
```

```python
#数据抽取
import pandas as pd 
data=pd.read_csv("data7.csv")
data.tel=data.tel.astype(str)  #转化字符串
data['bands']=data.tel.str.slice(0,3)#向量化运算，不能直接切片0-3位
data['areas']=data.tel.str.slice(3,7)
data['nums']=data.tel.str.slice(7,11)
data
```
