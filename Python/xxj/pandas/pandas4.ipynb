分布分析通常建立在定量分组的基础上，分布分析的要点：横轴也就是分组的

```python
data=pd.read_csv("data22.csv")
bins=[0,20,30,40,100]
agelabels=['<20',"21 30","31 40",">41",]
data['age']=pd.cut(data.年龄,bins,labels=agelabels)
data.groupby("年龄分层")['用户ID'].agg("count")
```

矩阵分析 是根据事物的两个重要属性或指标作为分析的一句进行关联分析，找出解决问题的一种分析方

```python
import pandas as pd
#矩阵分析实例
#CVR转化率 广告转化率
data = pd.read_csv('矩阵分析.csv')
#从消费和流量两个⻆度 分析各省的⽤户质量
costAgg = data.groupby(by= '省份', as_index =False)['⽉消费（元）'].agg('mean')
dataAgg = data.groupby(by= '省份', as_index =False)['⽉流量（MB）'].agg('mean')
aggData = costAgg.merge(dataAgg)
```

RFM分析是一个经典的商业数据模型 R代表近度 F代表频度 M代表金额，将数据进行分类

回归分析，是研究自变量和因变量之间数量变化关系的一种方法，通过建立因变量与自变量之间的回归模型，来预测因变量发展的趋势

回归分析与相关分析
相关分析和回归分析的联系：先进行相关分析，计算相关系数，再进行拟合回归模型，最后进行预测

相关分析和回归分析的区别：相关分析不分变量类型，且都是随机变量；回归分析一定要区分自变量和因变量，而且自变量是确定的普通变量



```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plot
from sklearn.linear_model import LinearRegression
mpl.rcParams['font.sans-serif'] = ['KaiTi']
# plot函数的汉字字体
mpl.rcParams['font.serif'] = ['KaiTi']

data = pd.read_csv('简单线性回归.csv')
x = data[['广告费用(万元)']]
y = data[['销售额(万元)']]
data['广告费用(万元)'].corr(data['销售额(万元)'])
data.plot('广告费用(万元)', '销售额(万元)',
kind='scatter')
plot.show()

lrModel = LinearRegression()
lrModel.fit(x, y) # 训练模型
print(lrModel.coef_) # 参数
print(lrModel.intercept_) # 截距
lrModel.score(x, y) # 拟合优度，决定系数
pX = pd.DataFrame(
{'广告费用(万元)': [20]}
)
lrModel.predict(pX)

lrModel = LinearRegression()
lrModel.fit(x, y) # 训练模型
print(lrModel.coef_) # 参数
print(lrModel.intercept_) # 截距
lrModel.score(x, y) # 拟合优度，决定系数
pX = pd.DataFrame(
{'广告费用(万元)': [20]}
)
lrModel.predict(pX)
```

```python
import pandas as pd
import matplotlib as mpl#设置中文字体
import matplotlib.pyplot as plot#绘图
from sklearn.linear_model import LinearRegression#机器学习库
mpl.rcParams['font.sans-serif'] = ['KaiTi']
# plot函数的汉字字体
mpl.rcParams['font.serif'] = ['KaiTi']
data = pd.read_csv('多重线性回归.csv')
x = data[['广告费用(万元)', '客流量(万人次)']]
y = data[['销售额(万元)']]
data['广告费用(万元)'].corr(data['销售额(万元)'])
data['客流量(万人次)'].corr(data['销售额(万元)'])

data.plot('广告费用(万元)', '销售额(万元)',kind='scatter')
data.plot('客流量(万人次)', '销售额(万元)',kind='scatter')


lrModel = LinearRegression()
lrModel.fit(x, y)
print(lrModel.coef_) # 参数
print(lrModel.intercept_) # 截距
lrModel.score(x, y) # 拟合优度，决定系数
pX = pd.DataFrame(
{'⼴告费用(万元)': [20], '客流量(万⼈次)': [5]}
)
lrModel.predict(pX)
```
